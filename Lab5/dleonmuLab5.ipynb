####Placing Orders


```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2017, 6, 1)
        self.SetEndDate(2017, 6, 15)
        
        
        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        self.iwm = self.AddEquity("IWM", Resolution.Minute)
        self.iwm.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("IWM",100)
            self.Debug(str(self.Portfolio["IWM"].AveragePrice))
        #4. Debug the AveragePrice of IWM
```

####Modified Algorithm



```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2017, 6, 1)
        self.SetEndDate(2017, 6, 15)
        
        
        #1,2. Select IWM minute resolution data and set it to Raw normalization mode
        self.iwm = self.AddEquity("AAPL", Resolution.Minute)
        self.iwm = self.AddEquity("TSL", Resolution.Minute)
        self.iwm = self.AddEquity("GOOG", Resolution.Minute)
        self.iwm = self.AddEquity("TRX", Resolution.Minute)
        self.iwm = self.AddEquity("ADA", Resolution.Minute)
        self.iwm.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
    def OnData(self, data):

        #3. Place an order for 100 shares of IWM and print the average fill price
        if not self.Portfolio.Invested:
            self.MarketOrder("AAPL",100)
            self.MarketOrder("TSL",100)
            self.MarketOrder("GOOG",100)
            self.MarketOrder("TRX",100)
            self.MarketOrder("ADA",100)
            self.Debug(str(self.Portfolio["AAPL"].AveragePrice))
            self.Debug(str(self.Portfolio["TSL"].AveragePrice))
            self.Debug(str(self.Portfolio["GOOG"].AveragePrice))
            self.Debug(str(self.Portfolio["TRX"].AveragePrice))
            self.Debug(str(self.Portfolio["ADA"].AveragePrice))
        #4. Debug the AveragePrice of IWM
```
